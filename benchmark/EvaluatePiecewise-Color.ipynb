```python
import numpy as np
import cv2
import numpy as np
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
import cv2

from freespacenn.config import Config, random_seed
from freespacenn.benchmark.data_loader_benchmarking import load_data
from freespacenn.util.vis import view_seg_map
from freespacenn.models.lenet import LeNet
```

```python
config = Config()
img, aux, label = load_data(config, seed=42, amount=50, resize=True)
```

```python
def slice_image(img):
    return np.array([np.vsplit(x, 7) for x in np.hsplit(img, 7)])

model = LeNet()
model.load_weights("/hdd/models/isef/lenet_patches_color_no_bn/model-0600.h5")

def infer_image(img_):
    img = img_ / 255.0
    img = 2 * (img - 0.5)
    slices = slice_image(img)
    slices = slices.reshape((49, 32, 32, 3))
    # inference
    out = model.predict(slices)
    out = out.reshape((7, 7))
    out = np.swapaxes(out, 0, 1)
    print(out)
    #out = np.around(out)
    draw = draw_on_image(img_, out)
    plt.imshow(draw, interpolation="none")
    cv2.imwrite("/hdd/vis/custom_sample.png", draw[:, :, ::-1])
    #print(out, out.shape)
    
from matplotlib.cm import ScalarMappable     

sm = ScalarMappable(cmap="jet")
sm.set_clim(0.0, 1.0)

def draw_on_image(img, out):
    scale0 = int(float(img.shape[0]) / float(out.shape[0]))
    scale1 = int(float(img.shape[1]) / float(out.shape[1]))
    
    spectrum = np.zeros((img.shape[0], img.shape[1] // 7, 3))
    spectrum[:, :] = np.kron(sm.to_rgba(np.linspace(0.0, 1.0, spectrum.shape[0]))[:, 0:3].reshape((spectrum.shape[0], 1, 3)),
                              np.ones((1, spectrum.shape[1], 1), dtype=out.dtype)) * 255.0
    spectrum = spectrum.astype(np.uint8)
    
    padding = np.zeros((img.shape[0], img.shape[1] // 14, 3), dtype=np.uint8)
    padding.fill(255)
    
    repeated = np.kron(out, np.ones((scale0, scale1), dtype=out.dtype))
    colors = sm.to_rgba(repeated)[:, :, 0:3]
    colors = np.array(colors * 255.0, dtype=np.uint8)
    r = img.copy()
    cv2.addWeighted(colors, 0.4, r, 0.6, 0, r)
    r = np.vstack([img, r])
    return r
```

```python
x = cv2.resize(cv2.imread("/hdd/vis/floordown.png")[:, :, ::-1], (224, 224))
infer_image(x) #(0.5 * (1.0 + img[40]) * 255.0).astype(np.uint8))
```

```python

def imginf(img_):
    img = img_ / 255.0
    img = 2 * (img - 0.5)
    slices = slice_image(img)
    slices = slices.reshape((49, 32, 32, 3))
    # inference
    out = model.predict(slices)
    out = out.reshape((7, 7))
    out = np.swapaxes(out, 0, 1)
    
    scale0 = int(float(img.shape[0]) / float(out.shape[0]))
    scale1 = int(float(img.shape[1]) / float(out.shape[1]))
    repeated = np.kron(out, np.ones((scale0, scale1), dtype=out.dtype))
    
    repeated = np.around(repeated)
    
    return repeated

y_pred = []
y_real = []
for i in range(len(img)):
    y_pred.append(imginf((0.5 * (1.0 + img[i]) * 255.0).astype(np.uint8)))
    y_real.append(label[i].argmax(axis=2))
    
y_pred = np.dstack(y_pred)
y_real = np.dstack(y_real)

print(y_pred.shape)
print(y_real.shape)

print("TP = {}".format(float(np.logical_and(y_real == 1, y_pred == 1).sum()) / np.prod(y_real.shape)))
print("TN = {}".format(float(np.logical_and(y_real == 0, y_pred == 0).sum()) / np.prod(y_real.shape)))
print("FP = {}".format(float(np.logical_and(y_real == 0, y_pred == 1).sum()) / np.prod(y_real.shape)))
print("FN = {}".format(float(np.logical_and(y_real == 1, y_pred == 0).sum()) / np.prod(y_real.shape)))
```
