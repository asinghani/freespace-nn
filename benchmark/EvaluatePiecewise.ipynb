```python
import numpy as np
import cv2
import numpy as np
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
import cv2

from freespacenn.config import Config, random_seed
from freespacenn.util.data_loader import load_data
from freespacenn.util.vis import view_seg_map
from freespacenn.models.lenet import LeNet
```

```python
config = Config()
train_data, test_data, valid_data, mean, stddev = load_data(config)

def get_img(gen):
    images, labels = next(gen)
    rgb, features, label = images[0][0], images[1][0], labels[0]
    
    image1 = (rgb[:, :, 0:3] / 2.0) + 0.5
    gt, overlay = view_seg_map(image1, label.argmax(axis=2), \
                      color=(0, 1, 0), alpha=0.6, include_mask=True)
    features = ((features / 2.0) + 0.5) * 255
    return (image1 * 255).astype(np.uint8), \
            (overlay * 255).astype(np.uint8), \
            (gt * 255).astype(np.uint8), \
        (cv2.resize(features[:, :, 0], None, fx = 2, fy = 2) \
            .reshape((224, 224))).astype(np.uint8), \
        (cv2.resize(features[:, :, 1], None, fx = 2, fy = 2) \
            .reshape((224, 224))).astype(np.uint8), \
        (cv2.resize(features[:, :, 2], None, fx = 2, fy = 2) \
            .reshape((224, 224))).astype(np.uint8)
```

```python
def slice_image(img):
    return np.array([np.vsplit(x, 7) for x in np.hsplit(img, 7)])

model = LeNet()
model.load_weights("/hdd/models/isef/lenet_patches_monochrome_no_bn/model-0666.h5")

def infer_image(img_, cmap=None):
    img = img_ / 255.0
    img = np.dot(img[...,:3], [0.2989, 0.5870, 0.1140])
    img = 2 * (img - 0.5)
    slices = slice_image(img)
    slices = slices.reshape((49, 32, 32, 1))
    # inference
    out = model.predict(slices)
    out = out.reshape((7, 7))
    out = np.swapaxes(out, 0, 1)
    #out = np.around(out)
    draw = draw_on_image(img_, out)
    plt.imshow(draw, interpolation="none", cmap=cmap)
    #print(out, out.shape)
```

```python
sm = ScalarMappable(cmap="jet")
sm.set_clim(0.0, 1.0)

def draw_on_image(img, out):
    scale0 = int(float(img.shape[0]) / float(out.shape[0]))
    scale1 = int(float(img.shape[1]) / float(out.shape[1]))
    
    spectrum = np.zeros((img.shape[0], img.shape[1] // 7, 3))
    spectrum[:, :] = np.kron(sm.to_rgba(np.linspace(0.0, 1.0, spectrum.shape[0]))[:, 0:3].reshape((spectrum.shape[0], 1, 3)),
                              np.ones((1, spectrum.shape[1], 1), dtype=out.dtype)) * 255.0
    spectrum = spectrum.astype(np.uint8)
    
    padding = np.zeros((img.shape[0], img.shape[1] // 14, 3), dtype=np.uint8)
    padding.fill(255)
    
    repeated = np.kron(out, np.ones((scale0, scale1), dtype=out.dtype))
    colors = sm.to_rgba(repeated)[:, :, 0:3]
    colors = np.array(colors * 255.0, dtype=np.uint8)
    r = img.copy()
    cv2.addWeighted(colors, 0.4, r, 0.6, 0, r)
    r = np.hstack([img, r, padding, spectrum])
    return r
```
