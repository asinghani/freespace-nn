```python
from freespacenn.util.vis import view_seg_map

from freespacenn.config import Config, random_seed
from freespacenn.util.data_loader import load_data
from freespacenn.models.floornet import FloorNet
```

```python
config = Config()
train_data, test_data, valid_data, mean, stddev = load_data(config)

def get_img(gen):
    images, labels = next(gen)
    rgb, features, label = images[0][0], images[1][0], labels[0]
    
    image1 = (rgb[:, :, 0:3] / 2.0) + 0.5
    gt, overlay = view_seg_map(image1, label.argmax(axis=2), \
                      color=(0, 1, 0), alpha=0.6, include_mask=True)
    features = ((features / 2.0) + 0.5) * 255
    return (image1 * 255).astype(np.uint8), \
            (overlay * 255).astype(np.uint8), \
            (gt * 255).astype(np.uint8), \
        (cv2.resize(features[:, :, 0], None, fx = 2, fy = 2) \
            .reshape((224, 224))).astype(np.uint8), \
        (cv2.resize(features[:, :, 1], None, fx = 2, fy = 2) \
            .reshape((224, 224))).astype(np.uint8), \
        (cv2.resize(features[:, :, 2], None, fx = 2, fy = 2) \
            .reshape((224, 224))).astype(np.uint8)
```

```python
config = Config()
config.init_weights = None
config.image_size = (224, 224)
config.input_shape = (config.image_size[0], config.image_size[1], 3)

model = FloorNet(config)
model.load_weights("/hdd/models/isef/sgd_carpetonly_nocrop/model-1999.h5")
```

```python
img_arr = []
aux_arr = []
lab_arr = []
for i in range(2):
    im, lab = next(train_data)
    img, aux = im
    for i in range(len(img)):
        img_arr.append(img[i].reshape((224, 224, 3)))
        aux_arr.append(aux[i].reshape((112, 112, 3)))
        lab_arr.append(lab[i].reshape((224, 224, 2)))
    
img_arr = np.array(img_arr)
aux_arr = np.array(aux_arr)
lab_arr = np.array(lab_arr)
```

```python
out = model.predict([img_arr, aux_arr])
```

```python
y_real = lab_arr.argmax(axis=3)
y_pred = out.argmax(axis=3)
```

```python
print("TP = {}".format(float(np.logical_and(y_real == 1, y_pred == 1).sum()))) # / np.prod(y_real.shape)))
print("TN = {}".format(float(np.logical_and(y_real == 0, y_pred == 0).sum()))) # / np.prod(y_real.shape)))
print("FP = {}".format(float(np.logical_and(y_real == 0, y_pred == 1).sum()))) # / np.prod(y_real.shape)))
print("FN = {}".format(float(np.logical_and(y_real == 1, y_pred == 0).sum()))) # / np.prod(y_real.shape)))
print("IOU = {}".format(float(np.logical_and(y_real == 1, y_pred == 1).sum()) / np.logical_or(y_real == 1, y_pred == 1).sum()))
```

```python
sess = tf.Session()

yt = tf.convert_to_tensor(out) #y_real.reshape((32, 224, 224, 1)))
yp = tf.convert_to_tensor(lab_arr) #y_pred.reshape((32, 224, 224, 1)))
score, up_opt = tf.metrics.mean_iou(yt, yp, 2)
sess.run(tf.local_variables_initializer())

with sess.as_default():
    print(sess.run(up_opt))
    print(sess.run(score))
```
