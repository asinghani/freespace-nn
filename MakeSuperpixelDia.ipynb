```python
import cv2
import numpy as np
from skimage.segmentation import slic
from skimage.segmentation import mark_boundaries
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("/hdd/floordown.png")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
superpixels = slic(img.astype(np.float64) / 255.0, n_segments = 200, sigma = 0)
```

```python
plt.imshow(mark_boundaries(img, superpixels))
```

```python
img1 = img.copy()
for i in range(108, 200):
    if i in [120, 125, 127, 134]:
        continue
    img1[(superpixels == i)] = (0, 255, 0)

cv2.addWeighted(img, 0.8, img1, 1 - 0.8, 0, img1)
cv2.imwrite("/hdd/vis/superpixels.png", mark_boundaries(img1, superpixels, color=(0.1, 0.1, 1))[:, :, ::-1] * 255)
cv2.imwrite("/hdd/vis/superpixels_1.png", mark_boundaries(img, superpixels, color=(0.1, 0.1, 1))[:, :, ::-1] * 255)

plt.imshow(mark_boundaries(img1, superpixels, color=(0.4, 0.4, 1)))
```
